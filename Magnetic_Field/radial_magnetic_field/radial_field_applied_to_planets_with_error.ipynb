```python
from astropy import constants as const
from astropy import units as u

# Constantes

r_s = const.R_sun.to(u.m).value.item() #UA
m_s= const.M_sun.to(u.kg).value.item() #kg
M_t= const.M_earth.to(u.kg).value.item() #kg
R_t= const.R_earth.to(u.m).value.item() #km
a=1.496e+11#m
```

```python
def campo_radial_RD(mi_0, sigma, U, Rd, Rdi, rho):
    
    Br=2*mi_0**(3/2)*sigma*U**2*(Rd-Rdi)**4*np.sqrt(rho)/(4*20*Rd**3) #Tesla

    return Br
```

```python
def campo_radial_superficie(mi_0, sigma, U, Rd, Rdi, rho, Rp):
    
    Br=2*mi_0**(3/2)*sigma*U**2*(Rd-Rdi)**4*np.sqrt(rho)/(20*4*Rp**3) #Tesla

    return Br
```

```python
def tesla_to_Gauss(Br):

    return Br*10**4
```

```python
def erro_campo(D, rp, rho, B, sigma_D, sigma_rho, sigma_rp):
    sigma=B*np.sqrt(4*(sigma_D/D)**2+(sigma_rho/(2*rho))**2+9*(sigma_rp/rp)**2)
    return sigma
```

```python
# GJ 1002 b
Rp=1.03*R_t
mi_0=4*np.pi*10**(-7) #permeabilidade magnética no vácuo
sigma=5*10**5 # condutividade eletrica do nucleo (ferro puro liquido)
Rd=0.526*Rp # raio externo do dinamo
Rdi=0.181*Rp # raio interno do dinamo
rho=11414.0 #densidade do material do nucleo (ferro puro). Difere da densidade propria do nucleo.
U=800/((Rd-Rdi)*mi_0*sigma)

B_r_sup=campo_radial_superficie(mi_0, sigma, U, Rd, Rdi, rho, Rp)
sigma_B_r_sup=erro_campo(Rd-Rdi, Rp/R_t, rho, B_r_sup, 2*0.043, 2*163, 0.2*Rp/R_t)
B_r_RD=campo_radial_RD(mi_0, sigma, U, Rd, Rdi, rho)
print(B_r_sup*10**4, sigma_B_r_sup*10**4, U)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['xtick.major.width'] = 1.0
plt.rcParams['ytick.major.width'] = 1.0
plt.rcParams['xtick.major.size'] = 4
plt.rcParams['ytick.major.size'] = 4

names = [p[0] for p in exoplanets]
B = np.array([p[1] for p in exoplanets])
sigma = np.array([p[2] for p in exoplanets])
x = np.arange(len(names))

def get_magnetosphere_color(B_val, sigma_val, Bmin_young, Bmin_current):
    threshold_current = min(Bmin_current) 
    threshold_young   = min(Bmin_young)   

    if B_val >= threshold_current:
        return "#2563eb"                        
    elif (B_val + sigma_val) >= threshold_current:
        return "#f97316"                  
    else:
        return "#6b7280"               

fig, ax = plt.subplots(figsize=(7.5, 5.5))
ax.set_facecolor('white')

ax.grid(axis="y", linestyle="--", alpha=0.3, linewidth=0.8, color='#cccccc', zorder=0)
ax.set_axisbelow(True)

current_min = min([min(p[4]) for p in exoplanets])
current_max = min(3.2, max([max(p[4]) for p in exoplanets]))

for i, p in enumerate(exoplanets):
    Bmin_young   = p[3]
    Bmin_current = p[4]

    ax.vlines(x=i, ymin=min(Bmin_young), ymax=max(Bmin_young),
              color="gray", linewidth=6, alpha=0.4, zorder=2)

    y_max_plot = min(3.2, max(Bmin_current))
    ax.vlines(x=i, ymin=min(Bmin_current), ymax=y_max_plot,
              color="blue", linewidth=6, linestyle=(0, (5, 3)), alpha=0.5, zorder=2)

    if max(Bmin_current) > 3.2:
        ax.annotate('', xy=(i, 3.15), xytext=(i, 3.0),
                   arrowprops=dict(arrowstyle='->', color='#1e40af', lw=1.5, alpha=0.6))

for i, p in enumerate(exoplanets):
    Bmin_young   = p[3]
    Bmin_current = p[4]
    color = get_magnetosphere_color(B[i], sigma[i], Bmin_young, Bmin_current)

    ax.errorbar(x[i], B[i], yerr=sigma[i],
                fmt="none", ecolor=color, elinewidth=2.0,
                capsize=4, capthick=2.0, alpha=0.8, zorder=4)

    ax.scatter(x[i], B[i], s=100, c='white',
               edgecolors=color, linewidths=2.5, zorder=5)
    ax.scatter(x[i], B[i], s=40, c=color, zorder=6)

for i in range(len(x)):
    y_pos = B[i] -0.025
    ax.text(i+0.45, y_pos, f'{B[i]:.2f}±{sigma[i]:.2f}',
           ha='center', va='bottom', fontsize=7,
           fontweight='normal', color='#333333',
           bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                    edgecolor='none', alpha=0.8))

ax.set_xticks(x)
ax.set_xticklabels(names, rotation=0, ha="center", fontsize=9, fontweight='bold')
ax.set_ylabel("Magnetic Field Strength (G)", fontsize=11, fontweight='bold')
ax.set_xlabel("Exoplanet", fontsize=11, fontweight='bold')
ax.set_ylim(-0.1, 3.2)
ax.set_xlim(-0.6, len(x) - 0.4)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.2)
ax.spines['bottom'].set_linewidth(1.2)

ax.text(0.5, 1.02, "Predicted Magnetic Fields vs. Magnetosphere Requirements",
       transform=ax.transAxes, fontsize=12, fontweight='bold',
       ha='center', va='bottom')

legend_items = [
    Line2D([0], [0], marker="o", color="w",
           markerfacecolor="#2563eb", markeredgecolor="#2563eb",
           markersize=8, markeredgewidth=2,
           label="Maintains current Earth magnetosphere"),
    Line2D([0], [0], marker="o", color="w",
           markerfacecolor="#6b7280", markeredgecolor="#6b7280",
           markersize=8, markeredgewidth=2,
           label="Maintains young Earth magnetosphere only"),
    Line2D([0], [0], marker="o", color="w",
           markerfacecolor="#f97316", markeredgecolor="#f97316",
           markersize=8, markeredgewidth=2,
           label="Marginally capable (B + σ ≥ current Earth req.)"),
    Line2D([0], [0], color="gray", lw=4, alpha=0.6,
           label="Young Earth requirement"),
    Line2D([0], [0], color="blue", lw=4, linestyle=(0, (5, 3)), alpha=0.6,
           label="Current Earth requirement"),
]

legend = ax.legend(handles=legend_items, loc='upper right',
                  fontsize=8, frameon=True, fancybox=False,
                  edgecolor='#cccccc', framealpha=0.95, ncol=1)
legend.get_frame().set_linewidth(1.0)

text_x, text_y = 1.5, 2.5
ax.text(text_x, text_y, 'Planets potential capable of maintaining\nCurrent Earth magnetosphere',
        fontsize=9, ha='center', color='black',
        bbox=dict(boxstyle='round,pad=0.4', facecolor='#e0f2fe',
                 edgecolor='black', linewidth=1),
        zorder=10)

for i in range(3):
    rad = 0.2 if i < 2 else -0.2
    ax.annotate('',
               xy=(i, B[i]),
               xytext=(text_x, text_y - 0.075),
               arrowprops=dict(arrowstyle='->',
                              connectionstyle=f'arc3,rad={rad}',
                              color='black', lw=1.5),
               zorder=9)

plt.tight_layout()
plt.savefig("magnetic_fields_nature_quality.png", dpi=600, bbox_inches='tight',
           facecolor='white', edgecolor='none')
plt.savefig("magnetic_fields_nature_quality.pdf", bbox_inches='tight',
           facecolor='white', edgecolor='none')
plt.show()
```
