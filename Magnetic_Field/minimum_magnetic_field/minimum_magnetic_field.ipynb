```python
from astropy import constants as const
from astropy import units as u

# Constantes
h = 6.626e-34   # J·s
c = 3e8         # m/s
kB = 1.381e-23  # J/K
m_s= const.M_sun.to(u.kg).value.item() #kg
T_sol=5780 #K
M_t= const.M_earth.to(u.kg).value.item() #kg
R_t= const.R_earth.to(u.km).value.item() #km
P_t= 1 #atm
UA=1.496e+11#m
R_sol=6.96*10**8 #m
```

```python
def campo_mag_max(Re, Te, Me, rsol, tsol, msol):
    B_max=4800*(Me*rsol**3*Te**8/(msol*Re**3*tsol**8))**(1/6)

    return B_max
    
```

```python
def campo_mag(Re, Te, Me, Rp, a, rm, rsol, tsol, msol):
    f=[0.002,0.005, 0.01, 0.02]
    #fração entre o Bzdi/Bkin, coloquei 0.25%, 0.5%, 1% e 2%, pq B_kin é aprox 4<B>. Estimativa conservadora. Obtive que em média B_kin=7.8<B> 
    B_m=[]
    for i in np.arange(len(f)):
        
        B_m.append(round(((3840*f[i]*Re**(3/2)/a**2)*(Me*rsol**3*Te**8/(msol*tsol**8))**(1/6)*(rm/Rp)**3), 2))

    return B_m
    
```

# Applying this to the studied exoplanets

```python
df_todos=pd.read_csv('exoplanets_potentially_habitable.csv')
```

```python
df_todos
```

```python
pl_name=df_todos["pl_name"].to_list()  
    

P_b=df_todos["pl_orbper (dias)"].to_list()   


a_p=df_todos["pl_orbsmax (UA)"].to_list()     

R_p=df_todos["pl_rade (R_t)"].to_list()   


M_p=df_todos["pl_bmasse (M_t)"].to_list()     


T_eff_p=df_todos["pl_eqt (K)"].to_list()    
   

T_e=df_todos["st_teff (K)"].to_list()      


R_e=df_todos["st_rad (R_sol)"].to_list()      
    

M_e=df_todos["st_mass (M_sol)"].to_list()       



```

```python
R_e[0]
```

```python
B_max=[]

for i in np.arange(len(R_e)):
    B_max.append(campo_mag_max(R_e[i]*R_sol, T_e[i], M_e[i]*m_s, R_sol, 5780, m_s))

B_min_5=[]

for i in np.arange(len(R_e)):
    B_min_5.append(campo_mag(R_e[i]*R_sol, T_e[i], M_e[i]*m_s, R_p[i], a_p[i]*UA, 5*R_p[i], R_sol, 5780, m_s))
B_min_11=[]

for i in np.arange(len(R_e)):
    B_min_11.append(campo_mag(R_e[i]*R_sol, T_e[i], M_e[i]*m_s, R_p[i], a_p[i]*UA, 11.7*R_p[i], R_sol, 5780, m_s))
```

```python
df_valor=pd.DataFrame({'pl_name':pl_name, 'B_max':B_max, 'B_min_5':B_min_5, 'B_min_11':B_min_11})
```
