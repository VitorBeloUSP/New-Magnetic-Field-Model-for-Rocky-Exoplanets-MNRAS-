```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(4, 2))

# Definir as cores e labels
materials = [
    ('Ice (<0)', '#A8D8EA'),
    ('Liquid water (0)', '#5DADE2'),
    ('Crust (1)', '#E8C4C4'),
    ('Olivine (2)', '#F9E79F'),
    ('Wadsleyite (3)', '#F8E8A8'),
    ('Ringwoodite (4)', '#E8B4B8'),
    ('Perovskite (5)', '#D4A5A5'),
    ('Post-Perovskite (6)', '#C4B5D8'),
    ('Outer core (7)', '#B8B8B8'),
    ('Inner core (8)', '#909090')
]

# Criar os patches para a legenda
legend_patches = []
for label, color in materials:
    patch = mpatches.Patch(facecolor=color, edgecolor='black', linewidth=0.5, label=label)
    legend_patches.append(patch)

# Adicionar a legenda
ax.legend(ncol=2, handles=legend_patches, loc='center', frameon=True, 
          fontsize=9, handlelength=2, handleheight=1)

# Remover os eixos
ax.axis('off')

# Ajustar o layout
plt.tight_layout()
plt.savefig('legenda.png')
# Mostrar a figura
plt.show()
```
